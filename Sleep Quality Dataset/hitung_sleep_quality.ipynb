```python
df = pd.read_excel("scale_quesioner_rename.xlsx", sheet_name="Lembar1")
df
```

```python
columns_to_new = [
    "ID",
    "C1",
    "C2",
    "C3",
    "C4_Q1",
    "C4_Q2",
    "C4_Q3",
    "C5_b",
    "C5_c",
    "C5_d",
    "C5_e",
    "C5_f",
    "C5_g",
    "C5_h",
    "C5_i",
    "C5_j",
    "C6",
    "C7_Q1",
    "C7_Q2"
]

df_use =df.loc[:, columns_to_new]
df_use
```

```python
df['C3'].describe()
```

```python
# Fungsi untuk mengonversi jawaban menjadi skor
def calculate_score(response):
    if response > 7:
        return 0
    elif response >= 6:
        return 1
    elif response >= 5:
        return 2
    else:
        return 3


df_use['C3_type'] = df_use['C3'].astype(str)

# Mengonversi string menjadi tipe data waktu (timedelta)
df_use['C3_type'] = pd.to_timedelta(df_use['C3_type'])
df_use['C3_hours'] = df_use['C3_type'].dt.components.hours

# Menggunakan metode apply untuk menerapkan fungsi ke setiap nilai dalam kolom 'C3'
df_use['Component 3 Score'] = df_use['C3_hours'].apply(calculate_score)

# Menampilkan DataFrame hasil
df_use
```

```python
# Konversi kolom waktu menjadi tipe data datetime
df_use['C4_Q2'] = pd.to_datetime(df_use['C4_Q2'], format='%H:%M:%S')
df_use['C4_Q3'] = pd.to_datetime(df_use['C4_Q3'], format='%H:%M:%S')

# Tambahkan satu hari ke waktu selesai jika waktu selesai lebih kecil dari waktu mulai
df_use.loc[df_use['C4_Q3'] < df_use['C4_Q2'], 'C4_Q3'] += pd.Timedelta(days=1)

# Hitung selisih waktu
df_use['selisih_waktu'] = df_use['C4_Q3'] - df_use['C4_Q2']

# Konversi selisih waktu menjadi jam
df_use['jam_kasur'] = df_use['selisih_waktu'].dt.total_seconds() / 3600

# Tampilkan DataFrame
df_use

```

```python
def calculate_score(row):
    tidur_kasur_ratio = row['C3_hours'] / row['jam_kasur']
    if tidur_kasur_ratio > 0.85:
        return 0
    elif tidur_kasur_ratio > 0.75:
        return 1
    elif tidur_kasur_ratio > 0.65:
        return 2
    else:
        return 3

# Menerapkan fungsi ke setiap baris dalam DataFrame
df_use["Component 4 Score"] = df_use.apply(lambda row: calculate_score(row), axis=1)
df_use
```

```python
# Custom function to calculate sum and assign scores
def calculate_sum_and_score(row):
    sum_C5 = sum(row['C5_b':'C5_j'])
    if sum_C5 == 0:
        return 0
    elif 1 <= sum_C5 <= 9:
        return 1
    elif 10 <= sum_C5 <= 19:
        return 2
    elif 20 <= sum_C5 <= 27:
        return 3
    else:
        return None  # Return None for undefined cases

# Apply the custom function to each row
df_use['Commponent 5 Score'] = df_use.apply(calculate_sum_and_score, axis=1)

df_use
```

```python
df_use['C6'].rename("Component 6 Score")
df_use
```

```python
# Custom function to calculate sum and assign scores
def calculate_sum_and_score_q7(row):
    sum_C7 = sum(row['C7_Q1':'C7_Q2'])
    if sum_C7 == 0:
        return 0
    elif 1 <= sum_C7 <= 2:
        return 1
    elif 3 <= sum_C7 <= 4:
        return 2
    elif  5<= sum_C7 <= 6:
        return 3
    else:
        return None  # Return None for undefined cases

# Apply the custom function to each row
df_use['Commponent 7 Score'] = df_use.apply(calculate_sum_and_score_q7, axis=1)

df_use
```

```python
# Calculate the total score
df_use["Total"] = df_use[['C1', 'C2', 'Component 3 Score', 'Component 4 Score', 'Commponent 5 Score', 'C6', 'Commponent 7 Score']].sum(axis=1)

# Determine sleep quality based on the total score
df_use['sleep_quality'] = df_use['Total'].apply(lambda x: 1 if x > 5 else 0)
df_use
```

```python
# Count the occurrences of each value in the 'sleep_quality' column
sleep_quality_counts = df_use['sleep_quality'].value_counts()

# Get the count of a specific value (e.g., 1)
count_of_1 = sleep_quality_counts.get(1)  # Get the count of value 1, default to 0 if not found

print(count_of_1)
```
